# Import data sets

```python
from sklearn.datasets import load_digits
```

```python
print(load_digits())
ld = load_digits()
```

```python
type(ld) # its not data frame
```

```python
ld["data"]
```

```python
ld["target"]
```

```python
y = ld["target"]
x = ld["data"]
```

# Now split the data

```python
from sklearn.model_selection import train_test_split as tts
xtrain,xtest,ytrain,ytest = tts(x,y,test_size=0.2,random_state = 1)
```

# now build model i.e perceptron model

```python
from sklearn.linear_model import Perceptron
per = Perceptron(max_iter=100)
per.fit(xtrain,ytrain)
ypred = per.predict(xtest)
```

```python
# check accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(ypred,ytest))
```

```python
# lets use iter_max = 100
from sklearn.linear_model import Perceptron
per1 = Perceptron(max_iter=10000000,random_state=1)
per1.fit(xtrain,ytrain)
ypred1 = per.predict(xtest)
# check accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(ypred1,ytest))
```
